```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Road_surface_type', hue='Accident_severity', palette='Set2')
plt.title('Accident Severity by Road Surface Type')
plt.xlabel('Road Surface Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Weather_conditions', hue='Accident_severity', palette='coolwarm')
plt.title('Accident Severity by Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Light_conditions', hue='Accident_severity', palette='mako')
plt.title('Accident Severity by Light Conditions')
plt.xlabel('Light Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Types_of_Junction', palette='rocket')
plt.title('Accident Hotspots by Junction Type')
plt.xlabel('Junction Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Age_band_of_driver', hue='Accident_severity', palette='viridis')
plt.title('Accident Severity by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python

cause_severity = df.groupby(['Cause_of_accident', 'Accident_severity']).size().unstack(fill_value=0)
cause_severity.plot(kind='bar', stacked=True, figsize=(14,7), colormap='tab20c')
plt.title('Cause of Accident vs. Severity')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python

severity_counts = df['Accident_severity'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Accident Severity Distribution')
plt.show()
```

```python

age_counts = df['Age_band_of_driver'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.plot(age_counts.index, age_counts.values, marker='o', color='teal')
plt.title('Accident Counts by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
